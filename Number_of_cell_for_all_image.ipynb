```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

cells = []
# List of six image file names
image_files = ['/content/Systems_bio_project/output31.png', '/content/Systems_bio_project/output18.png', '/content/Systems_bio_project/output21.png','/content/Systems_bio_project/output29.png', '/content/Systems_bio_project/output38.png']
              

# Loop through each image file name
for filename in image_files:

    # Load image
    image = cv2.imread(filename)

    # Convert to grayscale
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Apply Gaussian blur
    blur = cv2.GaussianBlur(gray, (7,11), 0)

    # Apply Canny edge detection
    canny = cv2.Canny(blur, 1, 255, 3)

    # Dilate edges
    dilated = cv2.dilate(canny, (1,1), iterations=2)

    # Find contours
    (cnt, heirarchy) = cv2.findContours(dilated.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)

    # Draw contours on original image
    rgb = cv2.cvtColor(blur, cv2.COLOR_BGR2RGB)
    cv2.drawContours(rgb, cnt, -1, (0,255,0), 2)

    # Display image with contours
    plt.imshow(rgb)
    plt.show()

    # Print number of contours found
    print('Number of contours found: ', len(cnt))
    cells.append(len(cnt))
```

```python
cells
```

```python
# Plot cells
plt.plot(cells)
plt.plot(cells, '*')
plt.xlabel('Time')
plt.ylabel('Number of Stress Granules')
plt.title('Number of Stress Granules Vs Time')
plt.show()
```
